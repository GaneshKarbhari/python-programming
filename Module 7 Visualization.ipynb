```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data=pd.read_csv("weather.csv")
```

```python
data
```

```python
data=data[:7]
```

```python
#matplotlib
```

```python
#line chart
```

```python
plt.plot(data["Date"],data["Humidity3pm"])
plt.title("Weather-Line Plot")
plt.xlabel("Date")
plt.ylabel("Humidity")
```

```python
#scatterplot
```

```python
plt.scatter(data["Date"],data["Humidity3pm"])
plt.title("Weather-Scatter Plot")
plt.xlabel("Date")
plt.ylabel("Humidity")
```

```python
#bar plot
```

```python
plt.bar(data["Date"],data["Humidity3pm"])
plt.title("Weather-Bar Plot")
plt.xlabel("Date")
plt.ylabel("Humidity")
```

```python
#histogram
```

```python
plt.hist(data["Humidity3pm"])
plt.title("Weather-Hist Plot")
plt.xlabel("Date")
plt.ylabel("Humidity")
```

```python
#seaborn
```

```python
sns.lineplot(x=data["Date"],y=data["Humidity3pm"],data=data)
```

```python
sns.scatterplot(x=data["Date"],y=data["Humidity3pm"],data=data)
```

```python
sns.barplot(x=data["Date"],y=data["Humidity3pm"],data=data)
```

```python
sns.histplot(y=data["Humidity3pm"],data=data)
```
